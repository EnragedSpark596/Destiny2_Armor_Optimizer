```python
# The calculator engine

# Select number of iterations

nIter = 50

for i in range(1,nIter+1):
    
    # Make a set of randomly selected armor
    helmResult = rn.choice(helms)
    gauntResult = rn.choice(gauntlets)
    chestResult = rn.choice(chestArmor)
    legResult = rn.choice(legs)
    classResult = rn.choice(classItems)

    passResult = [helmResult, gauntResult, chestResult, legResult, classResult]
    
    # Create buckets to store stat data
    typesList = []
    nameList = []
    energyList = []
    mobList = []
    resList = []
    recList = []
    discList = []
    intList = []
    strList = []
    
    # compile data into dataframe, ready for summation
    for dct in passResult:
        typesList.append(dct['Type'])
        nameList.append(dct['Name'])
        energyList.append(dct['Energy'])
        mobList.append(dct['Mob'])
        resList.append(dct['Res'])
        recList.append(dct['Rec'])
        discList.append(dct['Disc'])
        intList.append(dct['Int'])
        strList.append(dct['Stgth'])
        
    passData = {'Type':typesList,'Name':nameList,'Energy':energyList,'Mob':mobList,
                'Res':resList,'Rec':recList,'Disc':discList,'Int':intList,'Str':strList}
    
    # Create dataframe for current pass
    df_pass = pd.DataFrame.from_dict(passData)
    
    # Insert totals line at bottom of DF
    df_pass.loc[5] = ['Totals','N/A',df_pass['Energy'].sum(),df_pass['Mob'].sum(),df_pass['Res'].sum(),
                  df_pass['Rec'].sum(),df_pass['Disc'].sum(),df_pass['Int'].sum(),df_pass['Str'].sum()]
    
    #Outputs
    print(df_pass)
    
    print(np.sum(list(df_pass.loc[5])[3:]))
    print(np.mean(list(df_pass.loc[5])[3:]))
    print(np.std(list(df_pass.loc[5])[3:]))
```
